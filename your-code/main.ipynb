#### 1. Find the secret files.

```python
#when we want to look for files inside a repository we have to change the URL to a new one specifying that we want 
#to look fr the files with this structure : ****.scavengerhunt. In this case, we will look for the keyword "scavengerhunt"
#since this is the common word all wanted files have.

#More information of the method used can be found here: https://developer.github.com/v3/search/#search-code

end_point ='https://api.github.com/search/code?q=scavengerhunt+in:path+repo:ironhack-datalabs/scavenger'
headers={'Authorization': 'AIzaSyDeDEXWk7bq0r0GYzz-Pmc6dWqRGVxJ2Zc'}

request = requests.get(end_point)
```

#### 3. Read the content of each secret files into an array of strings.

```python
secret_files_df.head()
```

```python
# source where the code is found: https://developer.github.com/v3/repos/contents/
# Since the information of the files is provided but encrypted, a header has to be provided in order to 
#be able to read the files and be able to understand the data. 

header={'Accept': 'application/vnd.github.v3.raw'}

total_content=[]

for path in secret_files_df['path']: 
    print(path)
    end_point = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/'+ path
    request = requests.get(end_point, headers = header)
    print(request)
    total_content.append(request.text)

   #print(end_point)
#GET /repos/:owner/:repo/contents/:path
```
